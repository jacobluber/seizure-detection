```python
import os
import shutil
import numpy as np
import torch
latent_space=256
trainLatentPath='/home/axh5735/projects/signal_compression/logs/compression_128_256/tb_logs/compression_128_256/version_1/train_latent_spaces/'
validLatentPath='/home/axh5735/projects/signal_compression/logs/compression_128_256/tb_logs/compression_128_256/version_1/val_latent_spaces/'
testLatentPath='/home/axh5735/projects/signal_compression/logs/compression_128_256/tb_logs/compression_128_256/version_2/test_latent_spaces/'
gts=os.listdir('/home/axh5735/projects/vp_project/downstream_task/GT_new')
xTrain,xTest,xValid=np.empty((1,1,latent_space,26),dtype="float32"),np.empty((1,1,latent_space,26),dtype="float32"),np.empty((1,1,latent_space,26),dtype="float32")
yTrain,yTest,yValid=np.empty((1,1),dtype='float32'),np.empty((1,1),dtype='float32'),np.empty((1,1),dtype='float32')
fTrain,fTest,fVal=0,0,0
for gt in gts:
    #print(gt)
    gtPath='/home/axh5735/projects/vp_project/downstream_task/GT_new/'+gt
    l=gt.split('_')
    sample=l[0]+'_'+l[1]+'_'+l[4][:-4]+'_1.pt'
    if os.path.isfile(trainLatentPath+sample):
        print('hello')
        if fTrain==0:
            yTrain=np.load(gtPath).reshape((1,1))
            fTrain=1
            data=[]
            for i in range(1,27):
                #  code of xdata load
                tmp= trainLatentPath+l[0]+'_'+l[1]+'_'+l[4][:-4]+'_'+str(i)+'.pt'
                with open(tmp, 'rb') as filehandle:
                    transformation = torch.load(filehandle,map_location='cpu') #1
                numpy_array = transformation.cpu().detach().numpy()
                data.append(numpy_array)
            xTrain=np.array(data).reshape((1,1,latent_space,26)) 
                
        else:
            yTrain=np.vstack((yTrain,np.load(gtPath).reshape((1,1))))
            data=[]
            for i in range(1,27):
                tmp= trainLatentPath+l[0]+'_'+l[1]+'_'+l[4][:-4]+'_'+str(i)+'.pt'
                with open(tmp, 'rb') as filehandle:
                    transformation = torch.load(filehandle,map_location='cpu') #1
                numpy_array = transformation.cpu().detach().numpy()
                data.append(numpy_array)
            tmpData=np.array(data).reshape(1,1,latent_space,26)
            xTrain=np.vstack((xTrain,tmpData))

    
    elif os.path.isfile(validLatentPath+sample):
        if fVal==0:
            fVal=1
            yValid=np.load(gtPath).reshape((1,1))
            data=[]
            for i in range(1,27):
                tmp= validLatentPath+l[0]+'_'+l[1]+'_'+l[4][:-4]+'_'+str(i)+'.pt'
                with open(tmp, 'rb') as filehandle:
                    transformation =torch.load(filehandle,map_location='cpu') #1
                numpy_array = transformation.cpu().detach().numpy()
                data.append(numpy_array)
            xValid=np.array(data).reshape((1,1,latent_space,26))        
        else:
            yValid=np.vstack((yValid,np.load(gtPath).reshape((1,1))))
            data=[]
            for i in range(1,27):
                tmp= validLatentPath+l[0]+'_'+l[1]+'_'+l[4][:-4]+'_'+str(i)+'.pt'
                with open(tmp, 'rb') as filehandle:
                    transformation =torch.load(filehandle,map_location='cpu') #1
                numpy_array = transformation.cpu().detach().numpy()
                data.append(numpy_array)
            tmpData=np.array(data).reshape(1,1,latent_space,26)
            xValid=np.vstack((xValid,tmpData))


    elif os.path.isfile(testLatentPath+sample):
        if fTest==0:
            fTest=1
            yTest=np.load(gtPath).reshape((1,1))
            data=[]
            for i in range(1,27):
                tmp= testLatentPath+l[0]+'_'+l[1]+'_'+l[4][:-4]+'_'+str(i)+'.pt'
                with open(tmp, 'rb') as filehandle:
                    transformation =torch.load(filehandle,map_location='cpu') #1
                numpy_array = transformation.cpu().detach().numpy()
                data.append(numpy_array)
            xTest=np.array(data).reshape((1,1,latent_space,26)) 
        else:
            yTest=np.vstack((yTest,np.load(gtPath).reshape((1,1))))
            data=[]
            for i in range(1,27):
                tmp= testLatentPath+l[0]+'_'+l[1]+'_'+l[4][:-4]+'_'+str(i)+'.pt'
                with open(tmp, 'rb') as filehandle:
                    transformation =torch.load(filehandle,map_location='cpu') #1
                numpy_array = transformation.cpu().detach().numpy()
                data.append(numpy_array)
            tmpData=np.array(data).reshape(1,1,latent_space,26)
            xTest=np.vstack((xTest,tmpData))


yTrain,yTest,yValid=yTrain.astype(np.float32),yTest.astype(np.float32),yValid.astype(np.float32)
np.save('/home/axh5735/projects/signal_compression/task/256/onchip'+'/xTrain.npy',xTrain)
np.save('/home/axh5735/projects/signal_compression/task/256/onchip'+'/xTest.npy',xTest)
np.save('/home/axh5735/projects/signal_compression/task/256/onchip'+'/xVal.npy',xValid)
np.save('/home/axh5735/projects/signal_compression/task/256/onchip'+'/yTrain.npy',yTrain)
np.save('/home/axh5735/projects/signal_compression/task/256/onchip'+'/yTest.npy',yTest)
np.save('/home/axh5735/projects/signal_compression/task/256/onchip'+'/yVal.npy',yValid)
```
